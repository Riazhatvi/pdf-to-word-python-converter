```python
import fitz
from docx import Document
from docx.shared import Inches
from pathlib import Path
import tempfile
import shutil

# ==== FILE PATHS ====
pdf_path = "/content/file.pdf"
output_docx = "/content/drive/MyDrive/output_exact.docx"

print("Opening PDF...")

pdf = fitz.open(pdf_path)
doc = Document()

temp_dir = tempfile.mkdtemp()

try:
    for page_index in range(len(pdf)):

        page = pdf.load_page(page_index)

        # render page as image
        pix = page.get_pixmap(matrix=fitz.Matrix(2,2))

        img_path = f"{temp_dir}/page_{page_index}.png"
        pix.save(img_path)

        # insert image into Word
        doc.add_picture(img_path, width=Inches(6))

        if page_index < len(pdf)-1:
            doc.add_page_break()

        print(f"Processed page {page_index+1}/{len(pdf)}")

    doc.save(output_docx)

finally:
    shutil.rmtree(temp_dir)

print("✅ Conversion finished")
print("Download file from:", output_docx)
```
